```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_moons
import seaborn as sns

from mlxtend.plotting import plot_decision_regions

import tensorflow
from tensorflow.keras.models import Sequential

from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout

from tensorflow.keras.optimizers import Adam
```

```python
x, y = make_moons(100, noise = 0.25, random_state = 2)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x[:,0], x[:,1], c = y)
plt.show()
```

```python
model = Sequential()

model.add(Dense(128, activation='relu', input_dim = 2))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))

model.summary()
```

```python
adam = Adam(learning_rate=0.01)

model.compile(loss = 'binary_crossentropy', optimizer=adam, metrics = ['accuracy'])

history = model.fit(x,y, epochs=2000, validation_split=0.20, verbose = 0)
```

```python
plot_decision_regions(x, y.astype('int'), clf = model, legend =2)
plt.xlim(-2, 3)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
model = Sequential()

model.add(Dense(128, activation='relu', input_dim = 2, kernel_regularizer=tensorflow.keras.regularizers.l2(0.05)))
model.add(Dense(128, activation = 'relu', kernel_regularizer=tensorflow.keras.regularizers.l2(0.03)))
model.add(Dense(1, activation = 'sigmoid'))

model.summary()
```

```python
adam = Adam(learning_rate=0.01)

model.compile(loss = 'binary_crossentropy', optimizer=adam, metrics = ['accuracy'])

history = model.fit(x,y, epochs=2000, validation_split=0.20, verbose = 0)
```

```python
plot_decision_regions(x, y.astype('int'), clf = model, legend =2)
plt.xlim(-2, 3)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```
