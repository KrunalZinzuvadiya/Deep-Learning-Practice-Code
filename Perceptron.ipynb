```python
import pandas as pd
df = pd.read_csv('Datasets/placement.csv')
```

```python
df.head()
```

```python
x = df.iloc[:, 0:2]
y = df.iloc[:, -1]
```

```python
from sklearn.linear_model import Perceptron
```

```python
p = Perceptron()
```

```python
p.fit(x, y)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.values, y.values, clf = p, legend = 2)
```
